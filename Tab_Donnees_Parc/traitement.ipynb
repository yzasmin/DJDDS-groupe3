```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sélection des colonnes quantitatives pertinentes pour l'étude
quantitative_columns = ['Nb arbres', 'Occ_Nat', 'Occ_Ag', 'Occ_Ugs', 'Occ_eq', 'Occ_Tud', 'Occ_Tuc', 
                        'Surface', 'Argile', 'Limon', 'Sable', 'MO', 'N_Tot', 'Cu', 'Ni', 'Pb', 'Zn']


# Extraire les données pertinentes pour l'analyse
df_quant = df[quantitative_columns].dropna()

# Calculer la matrice de corrélation
correlation_matrix = df_quant.corr()

# Visualiser la matrice de corrélation
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Matrice de corrélation')
plt.show()

 
```

### Conclusion

La matrice de corrélation met en évidence plusieurs relations importantes entre les variables environnementales et la composition du sol. Tout d'abord, nous observons une forte corrélation positive entre la surface des parcelles et le nombre d'arbres, ce qui indique que les parcelles plus grandes tendent à avoir une densité d'arbres plus élevée. De plus, la forte corrélation entre la matière organique (MO) et l'azote total (N_Tot) montre que les sols riches en matière organique sont également riches en nutriments essentiels, ce qui pourrait favoriser la présence de vers de terre et de microorganismes. Concernant la texture du sol, il existe une corrélation négative marquée entre le sable et l'argile, indiquant que les parcelles sont soit sableuses, soit argileuses, mais rarement les deux à la fois. Enfin, l'occupation naturelle du sol semble être associée à des sols plus argileux, un type de sol qui retient mieux l'humidité et qui pourrait ainsi être plus favorable à la biodiversité du sol. Ces observations offrent des pistes pour comprendre les facteurs qui influencent la biodiversité dans les différentes parcelles étudiées.


***Effectuer l'ACP***

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardisation des données
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_quant)

# Application de l'ACP avec 5 composantes
pca = PCA(n_components=5)
pca_result = pca.fit_transform(df_scaled)

# Variance expliquée par chaque composante
explained_variance = pca.explained_variance_ratio_
print(f'Variance expliquée par les 5 premières composantes : {explained_variance}')
```

***Visualisation des resultats d'ACP***

```python
# Visualiser les résultats de l'ACP avec un graphique des deux premières composantes
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c='blue', edgecolor='k')
plt.title('Projection des données sur les deux premières composantes principales')
plt.xlabel(f'Composante 1 ({explained_variance[0]*100:.2f}% variance expliquée)')
plt.ylabel(f'Composante 2 ({explained_variance[1]*100:.2f}% variance expliquée)')
plt.grid(True)
plt.show()
```

### Conclusion

Le graphique ci-dessus montre la projection des données sur les deux premières composantes principales obtenues par l'Analyse en Composantes Principales (ACP). Ces deux composantes expliquent respectivement **21.63%** et **16.29%** de la variance totale, capturant donc environ **37.92%** de l'information contenue dans les variables d'origine. Cette réduction dimensionnelle permet de visualiser la répartition des parcelles dans l'espace des composantes principales. Bien que ces deux composantes n'expliquent pas toute la variance, elles montrent une dispersion intéressante des points, indiquant des différences importantes entre les parcelles. Les parcelles proches les unes des autres sur ce graphique partagent des caractéristiques similaires en termes de composition du sol, occupation du sol, ou autres variables environnementales, tandis que celles qui sont éloignées présentent des distinctions significatives dans ces mêmes caractéristiques. Cette visualisation constitue une étape préliminaire pour comprendre quels facteurs influencent le plus la biodiversité du sol, notamment la présence de vers de terre et de microorganismes.


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer
import pandas as pd
  
# Nettoyer les colonnes en enlevant les espaces avant et après dans les noms de colonnes
df.columns = df.columns.str.strip()

# Sélection des nouvelles colonnes pertinentes pour la régression
X = df[['Argile', 'Limon', 'Sable', 'Cu', 'Ni', 'Pb', 'Zn', 'Surface', 'Nb arbres', 'N_Tot', 
        'Occ_Nat', 'Occ_Ag', 'Occ_Ugs', 'Occ_eq', 'Occ_Tud', 'Occ_Tuc']]
y = df['MO']  # Variable cible (matière organique)

# Créer un objet pour imputer les valeurs manquantes (par la moyenne)
imputer = SimpleImputer(strategy='mean')

# Appliquer l'imputation sur les variables explicatives (X)
X_imputed = imputer.fit_transform(X)

# Diviser les données en ensembles d'entraînement et de test (80% entraînement, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Créer un modèle de régression linéaire
model = LinearRegression()

# Entraîner le modèle sur les données d'entraînement
model.fit(X_train, y_train)

# Faire des prédictions sur les données de test
y_pred = model.predict(X_test)

# Afficher les coefficients du modèle
print("Coefficients du modèle :", model.coef_)

 

# Évaluer le modèle
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Erreur quadratique moyenne (MSE) : {mse}")
print(f"Coefficient de détermination (R²) : {r2}")
```

Le modèle de régression linéaire montre que certaines variables, ont un impact significatif sur la matière organique (MO). En effet, N_Tot se distingue par une influence particulièrement importante. Le coefficient de détermination R² de 0.757 indique que 75.7% de la variance de la matière organique est expliquée par les variables sélectionnées, ce qui démontre une bonne performance du modèle. Cependant, l'erreur quadratique moyenne (MSE) de 73.95 suggère qu'il reste une marge d'amélioration.
