```python
import pandas as pd
#import du fichier excel
data = pd.read_excel('DataVdt.xlsx', sheet_name='DataRetravaille')
#afficher les 5 premieres lignes
data.head()
```

```python
# Appliquer la binarisation
data[['nb_read_avant_bioinfo', 'nb_read_apres_bio_info']] = data[['nb_read_avant_bioinfo', 'nb_read_apres_bio_info']].applymap(lambda x: 1 if pd.notnull(x) and x > 0 else 0)

# Afficher les premières lignes
data.head()
```

```python
data.to_excel('DataVdt.xlsx', index=False, sheet_name='DataRetravaille')
```
